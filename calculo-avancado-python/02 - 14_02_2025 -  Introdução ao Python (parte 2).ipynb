Estrutura de repetição

Contador => controla o loop

```python
#comentario

i=1 #inicialização da variável contadora
while i<=5:
  print(i)
  i+=1 #incremento +1
print("acabou")
```

Acumulador

```python
soma = 0
i=1 #inicialização da variável contadora
while i<=5:
  print(i)
  soma+=i # acumulador
  i+=1 #incremento +1
print("acabou")
print("total acumulado: " + str(soma))


```

Entrada e com quantidade

```python
#comentario
soma = 0
qtdPos = 0
i=1 #inicialização da variável contadora

while i<=5:
  num = int(input("Digite: ")) # entrada 5x
  print(i)
  soma+=num #acumular o que o usuário digitar
  i+=1 #incremento +1
  if num > 0:
    qtdPos+=1

print("Acabou")
print("Soma: ", soma)
print("Quantidade de valores positivos: ", qtdPos)
```

10 valores para o usuário apresentar a media e quantos valores são pares

```python
i = 1
soma = 0
media = 0

par = 0
impar = 0

while i <= 10:
  num = int(input(f"Digite o valor {i}: "))
  soma += num
  media = (soma / (i)) # tira 1 pois o "i" aumenta o valor em 1
  i+=1 #incrementa

  if (num%2 == 0):
    print(f"o valor {i - 1} é par")
    par+= 1
  else:
    print(f"o valor {i - 1} é impar")
    impar+= 1

print("\nmedia final: ", media)
print(f"quantidade de numeros pares: {par}")
print(f"quantidade de numeros impares: {impar}")
```

Ler 10 valores e armazenar numa lista. Depois armazene os números pares numa lista chamada par

```python
lista=[]
listaPar=[]

print("Informe os 10 valores: ")
for i in range(0,10): #[0,10[
  lista.append(int(input(f"Digite o número de ordem {i}: ")))

  if (lista[i]%2 == 0):
    listaPar.append(lista[i])

print("\nLista de todos os números: ", lista)
print("Lista dos números pares: ", listaPar)
```
