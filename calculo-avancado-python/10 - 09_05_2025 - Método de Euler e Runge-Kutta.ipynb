```python
import numpy as np
import matplotlib.pyplot as plt

h0 = 10          # altura inicial (m)
k = -0.2         # taxa de drenagem
t_max = 5        # tempo total (horas)
dt = 1           # passo (h)

# Vetores para tempo e altura
t = np.arange(0, t_max + dt, dt)
h = np.zeros(len(t))
h[0] = h0

# Método de Euler
for i in range(1, len(t)):
    f = k * h[i - 1]
    h[i] = h[i - 1] + dt * f

print("Tempo (h)\tAltura (m)")
for i in range(len(t)):
    print(f"{t[i]:>5}\t\t{h[i]:.4f}")

plt.plot(t, h, marker='o', label='Euler')
plt.title("Altura da Água no Reservatório (Euler)")
plt.xlabel("Tempo (horas)")
plt.ylabel("Altura (m)")
plt.grid(True)
plt.legend()
plt.show()
```
