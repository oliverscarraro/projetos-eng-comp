# Código para calcular a série de Taylor

# Série de Taylor

```python
import numpy as np
import matplotlib.pyplot as plt

# Função real
def real_func(x):
    return np.sqrt(x)

# Aproximação de Taylor de 3ª ordem ao redor de a = 1
def taylor_approx(x):
    return 1 + 0.5*(x - 1) - (1/8)*(x - 1)**2 + (3/16)*(x - 1)**3

# Intervalo de x
x_vals = np.linspace(0.8, 1.2, 400)
y_real = real_func(x_vals)
y_taylor = taylor_approx(x_vals)

# Gráfico
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_real, label='√x (função real)', color='blue')
plt.plot(x_vals, y_taylor, label='Aproximação de Taylor (3ª ordem)', color='red', linestyle='--')
plt.scatter([1.1], [real_func(1.1)], color='blue', s=50)
plt.scatter([1.1], [taylor_approx(1.1)], color='red', s=50)
plt.title('Aproximação de √x por Série de Taylor em torno de x=1')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

# Série de MacLaurin

```python
import numpy as np
import matplotlib.pyplot as plt

# Valor de pi
pi = np.pi

# Função real seno
def real_sin(x):
    return np.sin(x)

# Aproximação de Maclaurin de terceira ordem para sin(x)
def maclaurin_sin(x):
    return x - (x**3)/6  # sin(x) ≈ x - x³/3!

# Intervalo de x próximo de 0 (pois Maclaurin é centrado em x = 0)
x_vals = np.linspace(-np.pi/2, np.pi/2, 400)
y_real = real_sin(x_vals)
y_maclaurin = maclaurin_sin(x_vals)

# Ponto de interesse: π/6
x_point = pi / 6
real_val = real_sin(x_point)
approx_val = maclaurin_sin(x_point)

# Gráfico
plt.figure(figsize=(8, 5))
plt.plot(x_vals, y_real, label='sin(x) (função real)', color='blue')
plt.plot(x_vals, y_maclaurin, label='Maclaurin 3ª ordem', color='red', linestyle='--')
plt.scatter([x_point], [real_val], color='blue', s=50, label=f'sin(π/6) ≈ {real_val:.4f}')
plt.scatter([x_point], [approx_val], color='red', s=50, label=f'aprox ≈ {approx_val:.4f}')
plt.title('Aproximação de sin(x) por Série de Maclaurin (3ª ordem)')
plt.xlabel('x (rad)')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
