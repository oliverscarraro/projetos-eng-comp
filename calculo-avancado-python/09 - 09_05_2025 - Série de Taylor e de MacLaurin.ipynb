# Série de Taylor para aproximação

```python
import math
import sympy as sp

# 1. Fazendo para uma aproximação específica: f(x) = sqrt(x), a = 1 e x = 1,1

# Variáveis
a = 1
varX = sp.Symbol('x')
x = 1.1
fx = sp.sqrt(varX)

# Derivando a raiz de x
dfa1 = sp.diff(fx, varX)
dfa2 = sp.diff(fx, varX, 2)
dfa3 = sp.diff(fx, varX, 3)

# Substituindo x por a
dfa1_calc = dfa1.subs(varX, a)
dfa2_calc = dfa2.subs(varX, a)
dfa3_calc = dfa3.subs(varX, a)

# Aplicando na fórmula
formula = a + ((dfa1_calc) * (x - a)) + ((dfa2_calc)/(math.factorial(2)) * math.pow((x-a), 2)) + ((dfa3_calc)/(math.factorial(3)) * math.pow((x-a), 3))
print(formula)
```

# Série de MacLaurin para aproximação

```python
# import math (removida pois não estamos sempre tratando com valores reais - alguns simbólicos)
import sympy as sp # Cálculo de derivadas e etc (também utilizado em partes que antes estava usando math)
import numpy as np # Outras funções matemáticas
import matplotlib.pyplot as plt # Desenho do gráfico

# 1. Fazendo para uma aproximação específica: f(x) = sin(x), a = 0 e x = pi/6 rad (30°)

# Variáveis
a = 0 # Fixo para Série de MacLaurin
varX = sp.Symbol('x')
x = 0.523598775 # 30° em rad
fx = sp.sin(varX)

# Derivando a raiz de x
dfa1 = sp.diff(fx, varX)
dfa2 = sp.diff(fx, varX, 2)
dfa3 = sp.diff(fx, varX, 3)

# Substituindo x por a
dfa1_calc = dfa1.subs(varX, a)
dfa2_calc = dfa2.subs(varX, a)
dfa3_calc = dfa3.subs(varX, a)

# Aplicando na fórmula
calculo_maclaurin_30graus = a + ((dfa1_calc) * (x)) + ((dfa2_calc)/(2) * x**2) + ((dfa3_calc)/(6) * x**3)
print(calculo_maclaurin_30graus)
```

```python
# @title
def grafico():
  # Gerando o gráfico
  x_valores_grafico = np.linspace(-1.5, 1.5, 400) # vetor de 400 partes espaçando o intervalo de -1.5 à 1.5 igualmente

  # Aplicando na fórmula - Criando uma função simbólica, onde 'x' não tem valor diretamente definido
  expressao_maclaurin_generica = a + ((dfa1_calc) * (varX)) + ((dfa2_calc)/(2) * varX**2) + ((dfa3_calc)/(6) * varX**3)

  funcao_maclaurin = sp.lambdify(varX, expressao_maclaurin_generica, 'numpy') # transforma a expressão em uma função matemática utilizável

  y_valores_grafico = funcao_maclaurin(x_valores_grafico) # pontos y da forma da serie. calculo da serie para cada valor de x apresentado no gráfico

  # Função real: seno de x
  y_seno_real = np.sin(x_valores_grafico)

  # Criando o gráfico
  plt.figure(figsize=(8,5)) # tamanho 800x500

  # Desenhar funções e adicionar suas legendas
  plt.plot(x_valores_grafico, y_seno_real, label='sin(x) (funcao real)', color = 'blue') #função seno real
  plt.plot(x_valores_grafico, y_valores_grafico, label='Maclaurin 3ª ordem', color= 'red', linestyle='--')

  # Desenhar ponto onde sen = 30° - observação: x = correspondente a 30°, y = valor do seno real e aproximado neste ângulo
  # além disso, adicionar suas legendas. Valores dos pontos arredondados para 4 casas decimais
  plt.scatter([x], [np.sin(x)], color = 'blue', s = 50, label=f'sin(π/6) ≈ {np.sin(np.pi/6):.4f}') # s = size = tamanho
  plt.scatter([x], [calculo_maclaurin_30graus], color = 'red', s = 50, label=f'aprox ≈ {calculo_maclaurin_30graus:.4f}') # ponto da aproximação

  # títulos
  plt.title('Aproximação de sin(x) por Série de Maclaurin (3ª ordem)')
  plt.xlabel('x (rad)')
  plt.ylabel('y')

  plt.legend() # ativar legenda do campo superior esquerdo
  plt.grid(True) # ativa a grade do gráfico
```

```python
grafico()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Data for Top 5 Customer Locations
customer_locations_data = {
    'Estado': ['São Paulo (SP)', 'Minas Gerais (MG)', 'Rio de Janeiro (RJ)', 'Rio Grande do Sul (RS)', 'Paraná (PR)'],
    'Clientes (%)': [40.42, 12.63, 12.11, 6.18, 5.78]
}
df_customer_locations = pd.DataFrame(customer_locations_data)

# Sort data for bar chart
df_customer_locations_sorted = df_customer_locations.sort_values(by='Clientes (%)', ascending=False)


# Create a bar chart
plt.figure(figsize=(12, 7))
bars = plt.bar(
    df_customer_locations_sorted['Estado'],
    df_customer_locations_sorted['Clientes (%)'],
    color=['#4c72b0','#55a868','#c44e52','#8172b2','#ccb974'] # Example colors
)
plt.xlabel('Estado', fontsize=12)
plt.ylabel('Clientes (%)', fontsize=12)
plt.title('Clientes por Estado (Top 5)', fontsize=16)
plt.xticks(rotation=45, ha="right") # Rotate labels to prevent overlap
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to make room for labels

# Add percentage labels on top of bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 0.5, f'{yval:.2f}%', ha='center', va='bottom', fontsize=10)


# Save the chart
plt.savefig('clientes_por_estado.png')

print("Gráfico de barras para Clientes por Estado gerado e salvo como 'clientes_por_estado.png'")


# Data for Payment Methods
payment_methods_data = {
    'Tipo de Pagamento': ['Cartão de crédito', 'Boleto', 'Voucher', 'Cartão de débito'],
    'Participação (%)': [78.34, 17.92, 2.37, 1.37]
}
df_payment_methods = pd.DataFrame(payment_methods_data)

# Sort data for bar chart
df_payment_methods_sorted = df_payment_methods.sort_values(by='Participação (%)', ascending=False)

# Create a horizontal bar chart
plt.figure(figsize=(10, 6))
bars = plt.barh(
    df_payment_methods_sorted['Tipo de Pagamento'],
    df_payment_methods_sorted['Participação (%)'],
    color=['#4c72b0', '#55a868', '#c44e52', '#8172b2'] # Example colors
)

plt.xlabel('Participação (%)', fontsize=12)
plt.ylabel('Tipo de Pagamento', fontsize=12)
plt.title('Participação por Tipo de Pagamento', fontsize=16)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.gca().invert_yaxis() # To display the largest bar at the top

# Add percentage labels on the right of the bars
for bar in bars:
    xval = bar.get_width()
    plt.text(xval + 0.5, bar.get_y() + bar.get_height()/2.0, f'{xval:.2f}%', ha='left', va='center', fontsize=10)

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout() # Adjust layout to make room for labels

# Save the chart
plt.savefig('participacao_tipo_pagamento_barras.png')

print("Gráfico de barras horizontais para Formas de Pagamento gerado e salvo como 'participacao_tipo_pagamento_barras.png'")
```
