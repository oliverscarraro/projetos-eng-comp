Crie um programa que solicite 3 números e faça a média

```python
print("= Calculador de média =")
n1 = float(input("Digite o primeiro número: "))
n2 = float(input("Digite o segundo número: "))
n3 = float(input("Digite o terceiro número: "))
media = ((n1 + n2 + n3) / 3)
print("A média da operação é: ", media)
```

Estrutura de seleção

```python
num = int(input("Digite um número: "))
if num==0:
  print("O número digitado foi zero")
else:
  print("O número digitado é diferente de zero")

```

```python
num = int(input("Digite um número para verificar se ele é par ou ímpar: "))

if num%2==0:
  print("O número digitado é par")
else:
  print("O número digitado é ímpar")
```
