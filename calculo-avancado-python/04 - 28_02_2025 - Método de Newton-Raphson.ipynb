```python
# Definição da equação: f(x) = x^2 -2
# Derivada da equação: f'(x) = 2x

x = 1.5 # Chute inicial
tol = 10**(-5) # Tolerância para o erro
max_iter = 100 # Número máximo de iterações

for i in range(max_iter):

    fx = x**2 -2 # atribui f(x)
    dfx = 2*x # atribui f'(x) calculada

    if dfx == 0: # Evita divisão por zero
        print("Derivada zerada! Escolha outro chute inicial")
        break

    x_new = x - fx / dfx # Fórmula de Newton-Raphson

    if abs(x_new - x) < tol: # Verifica o critério de parada
        print(f"Raiz encontrada: {x_new}")
        print(f"Convergiu em {i+1} iterações.")
        break

    x = x_new # Atualiza x para a próxima iteração

else:
    print("Numero máximo de iterações atingido. Solução pode não ser precisa.")
```
