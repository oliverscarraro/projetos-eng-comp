# **FUNÇÃO**

```python
# quando escrevemos "def", o código fonte compreende sabe que estamos criando uma função

def msg():
    print("Boa noite! Bem vindo a calculadora!")

def soma(a,b): # esses são os parâmetros
    soma = a+b
    return soma

def mult(a,b): # elas também são variáveis locais
    return a*b

def main(): # não é necessário no Python
    opcao = int(input("Escolha uma opção: "))
    if opcao == 1:
      msg
    elif opcao == 2:
      print("Você escolheu a soma: ")
      num1 = int(input("Digite um número: "))
      num2 = int(input("Digite outro número: "))
      print(soma(num1,num2))
    elif opcao == 3:
      print("Você escolheu a multiplicação: ")
      num1 = int(input("Digite um número: "))
      num2 = int(input("Digite outro número: "))
      # mesmo usando num1 e num2 ao invés de 'a' e 'b', funciona porque ele utiliza a posição
      print(mult(num1,num2))
    else:
      print("Opção incorreta!")

main()




```

Faça um programa que solicita ao usuário um número inteiro de 1 a 10 e verifique se ele acertou o número. Utilize a biblioteca 'random'

```python
import random as aleatorio

acertou = False
sorteado = aleatorio.randint(0,10)
escolhido = int(input("Digite um número de 0 a 10. Será avisado se foi o número sorteado: "))

while (acertou == False):
    if (escolhido == sorteado):
        print(f"Você acertou o número sorteado! Era o '{escolhido}'")
        acertou = True
    else:
        escolhido = int(input("Tente novamente: esse não era o número sorteado."))


```
